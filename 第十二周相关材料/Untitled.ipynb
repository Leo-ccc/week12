```python
import panel as pn
import seaborn as sns
```

```python
import hvplot.pandas
```

```python
df = sns.load_dataset('iris')
df
```

```python
df.hvplot.scatter(x = 'sepal_length', y = 'sepal_width', by = 'species')
```

```python
x = pn.widgets.Select(name = 'x', options=['sepal_width', 'petal_width'])
y = pn.widgets.Select(name = 'y', options=['sepal_length', 'petal_width'])
kind = pn.widgets.Select(name = 'kind', options=['scatter','bivariate'])
plot = df.hvplot(x=x,y=y,kind=kind,width = 600)
```

```python
pn.Row(pn.WidgetBox(x,y,kind),plot)
```
